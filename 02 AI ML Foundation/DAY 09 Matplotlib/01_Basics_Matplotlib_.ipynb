Data Visualization Using Matplotlib : Line Plot

```python
import matplotlib.pyplot as plt
```

```python
X = [1,2,3,4]
Y = [5,6,7,8]
plt.plot(X,Y)
plt.show()
```

```python
# Movies Data

# Oscar Winner Movies
oscar_movies = [
    "The Dark Knight", 
    "The Hurt Locker",  
    "The King's Speech", 
    "The Artist",     
    "Argo"
]
oscar_years = [2008, 2009, 2010, 2011, 2012]
oscar_revenue = [1005, 170, 427, 133, 232]  # in $M

# Plot
plt.plot(oscar_years,oscar_revenue)

# Methods
plt.title("Oscar Movies Revenue (From Year 2008 to 2012)")
plt.xlabel("Years")
plt.ylabel("Revenue (in $M)")
```

```python
# Plot multiple
# Non-Oscar Winner Movies
non_oscar_movies = ["Slumdog Millionaire","Avatar","Inception","Hugo","Lincoln"]
# non_oscar_years = [2008, 2009, 2010, 2011, 2012]
non_oscar_revenue = [378, 2788, 829, 185, 275]  # in $M

plt.plot(oscar_years ,non_oscar_revenue ,"bo--", label="Oscar Movies")
plt.plot(oscar_years ,oscar_revenue,color = "red" , linestyle=":", marker=">" ,label="Non-Oscar Movies" )
# plt.plot(oscar_years ,non_oscar_revenue ,label="Oscar Movies")
# plt.plot(oscar_years ,oscar_revenue,label="Non-Oscar Movies" )

plt.title("Oscar Movies VS Non Oscar Movies Revenue")
plt.xlabel("Years")
plt.ylabel("Revenue (in $M)")

plt.legend()
```
