Bar Plot

```python
import matplotlib.pyplot as plt
```

Vertical Bar Charts

```python
# Oscar Winner Movies
oscar_movies = [
    "The Dark Knight", 
    "The Hurt Locker",  
    "The King's Speech", 
    "The Artist",     
    "Argo"
]
oscar_years = [2008, 2009, 2010, 2011, 2012]
oscar_revenue = [1005, 170, 427, 133, 232]  # in $M
plt.bar(oscar_years ,oscar_revenue , color="green" )
plt.title("Revenue of Oscar Movies")
plt.xlabel("Years")
plt.ylabel("Oscar Movies")
for i in range(len(oscar_years)):
        plt.text(oscar_years[i] ,oscar_revenue[i] + 20 , oscar_revenue[i] , ha = "center"  )

plt.ylim(0,max(oscar_revenue) + 100)
plt.xlim(min(oscar_years) -1 , max(oscar_years) + 1)
```

Multiple DataSet on Bar Chart

```python
# comparing data with bar charts
oscar_movies = ["The Dark Knight","The Hurt Locker","The King's Speech","The Artist","Argo"]
oscar_revenue = [1005, 170, 427, 133, 232]  # in $M

non_oscar_movies = ["Slumdog Millionaire","Avatar","Inception","Hugo","Lincoln"]
non_oscar_revenue = [378, 2788, 829, 185, 275]  # in $M

years = [2008, 2009, 2010, 2011, 2012]

# Plotting Bar Chart
import numpy as np
x = np.arange(len(years))
width = 0.4

plt.bar(x-width/2, oscar_revenue, width, label="Oscar")
plt.bar(x+width/2, non_oscar_revenue, width, label="Non-Oscar")

plt.xlabel("Years")
plt.ylabel("Revenue (in $M)")
plt.title("Yearly revenue of Oscar vs Non-Oscar movies")
plt.legend()
plt.tight_layout()
```
