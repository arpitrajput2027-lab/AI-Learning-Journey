Melt and Pivot

```python
# Melt 
import pandas as pd
df = pd.DataFrame({
    "country": ["USA", "USA", "India", "India"],
    "year": [2020, 2021, 2020, 2021],
    "sales": [100, 120, 90, 110],
    "profit": [20, 25, 18, 22],
    "tax":[10,20,30,20]
})

melted_df = df.melt(
    id_vars = ["country","year"],
    value_vars = ["sales" , "profit","tax"],
    var_name = "combined_column",
    value_name = "Value"
)
melted_df
```

```python
#pivot
melted_df.pivot(
    index = ["country","year"], # paire - unique
    columns = "combined_column",
    values = "Value"
)
```
