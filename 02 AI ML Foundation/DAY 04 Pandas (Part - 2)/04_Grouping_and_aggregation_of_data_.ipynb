Grouping and Aggregation of the DATA

```python
import pandas as pd
df = pd.read_csv("raw_data.csv")
print(df)
```

```python
print(df.groupby("country")["income"].mean())
print()
print(df.groupby("country")["income"].min())
print()
print(df.groupby("gender")["income"].mean())
```

```python
print(df.groupby("country")["income"].agg(["mean","min","max"]))
print(df.groupby("country")["income"].aggregate(["mean","min","max"])) # same work as above

print()
print(df.groupby("country")["income"].agg(avg_income="mean",min_income="min",max_income="max"))
```

```python
print(df.groupby("country").agg({
    "income" : "mean",
    "age" : "max"
}))

print()

print(df.groupby("country").agg(
    avg_income = ("income","mean"),
    max_age = ("age" ,"max")
))
```
