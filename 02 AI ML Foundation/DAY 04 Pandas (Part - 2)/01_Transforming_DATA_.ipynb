Transforming DATA - apply() , map() , assign() , replace() ,rename(columns={old:new}) , sort_values() , sort_index() , reset_index() , rank() , Reorder Columns

```python
import pandas as pd
df = pd.read_csv("raw_data.csv")

df2 = df.copy()
df2["tax"] = df2["income"].apply(lambda x : "20%" if x >= 60_000  else "10%")
print(df2)
```

```python
gender_map = { "Male" : "M" , "Female" : "F","Unknown" : " - "}

df2["gender"] = df2["gender"].map(gender_map)
print(df2)
```

```python
df2 =  df2.assign( Incremented_Income = df2["income"]*1.1)
df2
```

```python
df2["country"] = df2["country"].replace("USA","United States of America")
df2
```

```python
df2.columns = ["ID","NAME","AGE","COUNTRY","GENDER","INCOME","TAX","NEW INCOME"]
df2
```

```python
df2.rename(columns = {"INCOME":"SALARY"})
df2.rename(index = {1:"First"})
```

```python
df2.sort_values("INCOME")
df2.sort_values("INCOME" , ascending=False)
df2 = df2.fillna(100)
sorted_df = df2.sort_values(["INCOME" , "AGE"]) # age will be  2nd Parameter for Sorting
sorted_df.sort_index()

# reset
# sorted_df.reset_index()
sorted_df.reset_index(drop=True)

```

```python
sorted_df["Ranking"] = sorted_df["INCOME"].rank(ascending = True , method = "dense") # method -> min or max
sorted_df
```
