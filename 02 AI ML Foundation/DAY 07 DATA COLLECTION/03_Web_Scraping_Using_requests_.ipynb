Web Scraping Using requests

```python
import requests
```

```python
URL = "https://www.scrapethissite.com/pages/simple/"
res = requests.get(URL)
```

```python
print(res)
```

```python
if res.status_code == 200:
    print("It is Good Practice to Firstly Check the Status Code then Excecute other things")
else :
    print("Error ion Fetching DATA.")

    
print(res.status_code) # HTTP status (200 = OK)
# print(res.text) # HTML content of the page as a string
# print(res.content) # Raw bytes of the response
# print(res.headers) # Metadata like content type, server info etc.print(res.status_code) # HTTP status (200 = OK)

```

Downloading & Storing to Files

```python
with open("scraped_data/data1.html" , "w" , encoding = "utf-8") as f:
    f.write(res.text)
```
