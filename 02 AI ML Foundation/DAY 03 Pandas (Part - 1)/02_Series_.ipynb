#Series...

```python
import pandas as pd

# Series
series = pd.Series(["Arpit Rajput" , "Aryan Deo" , "Bhavay Vij" , "Bhavneesh "])
print(series)

#index
print(series[0])
print(series[2])
```

```python
# labeled based indexing and accessing
ser = pd.Series(["17" , "18" , "19" , "20","X"] , index = ["Arpit Rajput" , "Aryan Deo" , "Bhavay Vij" , "Bhavneesh ","Unknown"])
print(ser)
print()
print(ser["Arpit Rajput"])
print(ser["Aryan Deo"])
print(ser[2])
```
