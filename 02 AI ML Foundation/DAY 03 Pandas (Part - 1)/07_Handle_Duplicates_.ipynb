Handle Missing Values in Given DATA SET

```python
import pandas as pd
df = pd.read_csv("07_employees_infos_1_.csv")

df.isnull() # Give True for Missing Values
df.isna() # same as  df.isnull()
df.isnull().sum() # Give the no of missing values of each column
df.dropna() # removes the rows which includes missing values
df.dropna(axis=1) # removes the column which includes missing values
```

```python
# fill the missing values 
df.fillna(69)

# age_mean = df["age"].mean() # but this will not reflect on Original DATA
# df["age"].fillna(age_mean)

cleaned_data = df.copy()
age_mean = cleaned_data["age"].mean() # cleaned_data is COPY of  Original DATA
cleaned_data["age"] = cleaned_data["age"].fillna(age_mean)

cleaned_data
```

```python
df.ffill() # Forward fill
df.bfill() # Backward fill
```
