Data frame - Creation

```python
import pandas as pd

# Dictionary of Lists
detail = {
    "name":["Arpit Rajput" , "Aryan Deo" , "Bhavay Vij" , "Bhavneesh "],
    "course": ["AI & ML", "DSA","Web Dev","Data Science"],
    "age": [19,20,19,20],
    "city": ["CHD","JRK","HR","PB"],
    "cgpa": [8.3,9.6,8.9,8.98]
}

data_frame = pd.DataFrame(detail)
print()
print(data_frame , type(data_frame))
print()
print(data_frame.index)
print()
print(data_frame.columns)
```

```python
# from List of Lists
df = pd.DataFrame([["Arpit",18],["Bhavay",19],["Aryan",20],["Bhavneesh",21]] , columns = ["Name","Age"])
print(df)
```

```python
# from numpy array
import numpy as np
np_arr = np.array([["Arpit",18,8.3],["Bhavay",19,8.89],["Aryan",20,9.6],["Bhavneesh",21,8.98]] )

data_frame = pd.DataFrame(np_arr ,  columns = ["Name","Age","CGPA"])
print(data_frame)
```
