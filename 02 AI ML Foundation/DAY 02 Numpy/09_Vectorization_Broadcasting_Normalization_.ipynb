Vectorization and Broadcasting

```python
import numpy as np
arr = np.array([1,2,3,4,5,6])

#Vectorization
print(arr**3)
print(arr+10)
```

```python
# Broadcasting

arr2 = np.array([[1,2,3],[4,5,6],[8,9,0]]) # (3,3) -> (1,3)
arr3 = np.array([4,5,6])
print(arr2 + arr3)
```

```python
# Normalization
arr4 = np.array([[1,2],[3,4]])
mean = np.mean(arr4)
std_dev = np.std(arr4)

normalized_arr = ((arr4-mean)/std_dev)
print(normalized_arr)
print(mean , std_dev)
```
