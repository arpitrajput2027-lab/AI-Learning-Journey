Pyhton List vs Numpy Array

```python
import numpy as np
import time
```

```python
#execution Performance
# Python List
size = 1000
py_list = list(range(size))
start = time.time()
[x*x for x in py_list]
end = time.time()
py_time = end - start
print(f"Time taken by Python List : {end - start}")

# numpy array
np_arr = np.array(py_list)
start = time.time()
np_arr**2
end = time.time()
np_time = end - start
print(f"Time taken by Python List : {end - start}\n")
print(f"Time Difference : {py_time - np_time}")
```

```python
import sys
print(f"Size of Python List : {sys.getsizeof(py_list)} bytes ")
print(f"Size of Numpy Array : {sys.getsizeof(np_arr)} bytes ")
print(f"Size of Numpy Array : {np_arr.nbytes} bytes ")
```

```python

print("List container size:", sys.getsizeof(py_list))
print("One int size:", sys.getsizeof(py_list[0]))
print("Approx total list memory:", sys.getsizeof(py_list) + size * sys.getsizeof(py_list[0]))

print("NumPy data size:", np_arr.nbytes)
```
