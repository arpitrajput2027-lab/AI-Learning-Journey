DATA TYPES

```python
import numpy as np

# int 64 , int 128
arr1 = np.array([1,2,6,7,8])
print(arr1.dtype)

# float64 float 128
arr2 = np.array([1,2,6,7,8,3.14])
print(arr2.dtype)

# U32 (Unicode) basically string
arr3 = np.array([1,2,6,7,8,3.14 ,"Venom"])
print(arr3.dtype)

# boolean
arr5= np.array([False , True , False,True ,(5<8)])
print(arr5.dtype)

# Complex
arr6 = np.array([2 + 6j])
arr7 = np.array([4 + 3j])
print(arr6 + arr7 , (arr6 + arr7).dtype)

# Object  - sab array mei sab hi aa jaye to vo direct object hi btata hai
arr8 = np.array([1 , "venom" , 3.14 , {3 , 4 ,5 ,6}])
print(arr8.dtype)
```
