Mathematical Funtions

```python
import numpy as np
# aggregation
arr = np.array([1,2,3,4,5])

print(np.sum(arr))
print(np.prod(arr))
print(np.max(arr))
print(np.min(arr))
print(np.argmax(arr))
print(np.argmin(arr))

print(np.mean(arr))
print(np.std(arr))
print(np.median(arr))
print(np.var(arr))
```

```python
# power fxn
arr = np.array([1,2,3,4,5])

print(np.square(arr))
print(np.sqrt(arr))
print(np.pow(arr , 3))
```

```python
arr = np.array([1,2,3,4,5,10,16])

# log fxns
print(np.log(arr))
print(np.log2(arr))
print(np.exp(arr))
```

```python
# rounding fxns

print(np.round(3.14739))
print(np.ceil(3.14739))
print(np.floor(3.14739))
print(np.trunc(3.14739))
print(np.round(3.14739))
```

```python
arr = np.array([1, 2, -5, 3, 8, -4, 2, 5])
print(np.abs(arr))     # [1 2 5 3 8 4 2 5]
print(np.sort(arr))    # [-5 -4  1  2  2  3  5  8]
print(np.unique(arr))  # [-5 -4  1  2  3  5  8]
```
